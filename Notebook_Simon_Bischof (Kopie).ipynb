# Dynamische Graphen generieren

Zuerst importieren wir NetworKit:

```python
from NetworKit import *
```

## Benutzen von statischen Graphgeneratoren

Die statischen Generatoren des Watts-Strogatz-Modells und Dorogotsev-Mendes-Modells werden analog zu den restlichen statischen Generatoren aufgerufen: Im Konstruktor werden die Parameter übergeben.

Der Watts-Strogatz-Generator nimmt als Paramter eine Anzahl Knoten, die Anzahl der Nachbarn auf jeder Seite eines Knotens im regulären Ringgitter und die Wahrscheinlichkeit für das Neuverdrahten von Kanten:

```python
WSG = generators.WattsStrogatzGenerator(20,3,0.05)
```

Der Dorogotsev-Mendes-Generator nimmt eine Anzahl Knoten als Parameter:

```python
DMG = generators.DorogovtsevMendesGenerator(20)
```

Die Generatoren haben eine Methode "generate()", die den Graphen generiert:

```python
GWS = WSG.generate()
viztasks.drawGraph(GWS)
```

```python
GDM = DMG.generate()
viztasks.drawGraph(GDM)
```

## Benutzen von dynamischen Graphgeneratoren

Die dynamischen Generatoren des Forest-Fire-Modells und Dorogotsev-Mendes-Modells werden analog zu den restlichen dynamischen Generatoren aufgerufen: Im Konstruktor werden die meisten Parameter übergeben.

Der Forest-Fire-Generator nimmt als Parameter einen "Flammbarkeitsparameter", der angibt, wie wahrscheinlich das Entflammen eines Knotens ist:

```python
DFFG = generators.DynamicForestFireGenerator(0.4)
```

Der Dorogotsev-Mendes-Generator nimmt keine Parameter:

```python
DDMG = generators.DynamicDorogovtsevMendesGenerator()
```

Die Methode "generate(count nSteps)" führt nSteps Zeitschritte aus. Zurückgegeben wird ein Stream, der die Änderungen am Graphen speichert:

```python
streamDDMG = DDMG.generate(17)
streamDFFG = DFFG.generate(27)
```

Durch erneutes Aufrufen von generate werden weitere Zeitschritte getan:

```python
streamDFFG2 = DFFG.generate(2)
```

Um daraus einen Graphen zu erstellen, kann die Klasse GraphUpdater verwendet werden:

```python
GDDM = Graph()
GUDDM = dynamic.GraphUpdater(GDDM)
GUDDM.update(streamDDMG)
viztasks.drawGraph(GDDM)
```

```python
GDFF = Graph()
GUDFF = dynamic.GraphUpdater(GDFF)
GUDFF.update(streamDFFG)
GUDFF.update(streamDFFG2)
viztasks.drawGraph(GDFF)
```

## Eigenschaften der Generatoren

Der Watts-Strogatz-Generator erstellt ein Netzwerk mit geringem Durchmesser, allerdings ist das Potenzgesetz für die Gradverteilung nicht erfüllt. Die Anzahl der Kanten ist nicht zufällig, sondern schon durch die Parameter vorbestimmt (nämlich das Produkt der ersten beiden Parameter):

```python
WSG = generators.WattsStrogatzGenerator(100,4,0.3)
GWS = WSG.generate()
properties.overview(GWS)
```

Das das Potenzgesetz nicht erfüllt ist, ist auch im folgenden Plot zu erkennen, wo sich sonst eine Linie ergeben hätte (da logarithmische Skalen):

```python
dd = properties.degreeDistribution(GWS)
xscale("log")
xlabel("degree")
yscale("log")
ylabel("number of nodes")
plot(dd)
```

Die Dorogovtsev-Mendes-Generatoren zeigen tatsächlich einen geringen Durchmesser und ein Potenzgesetz bei der Gradverteilung. Aber auch hier ist die Anzahl der Kanten nicht zufällig ($m = 2n-3$): 

```python
DMG = generators.DorogovtsevMendesGenerator(1000)
GDM = DMG.generate()
properties.overview(GDM)
```

```python
DDMG = generators.DynamicDorogovtsevMendesGenerator()
GDDM = Graph()
GUDDM = dynamic.GraphUpdater(GDDM)
GUDDM.update(DDMG.generate(997))
properties.overview(GDDM)
```

Im folgenden Diagramm ergibt sich eine ziemlich gute Linie:

```python
dd2 = properties.degreeDistribution(GDM)
xscale("log")
xlabel("degree")
yscale("log")
ylabel("number of nodes")
plot(dd2)
```

Der Forest-Fire-Generator zeigt ebenfalls einen geringen Durchmesser und ein Potenzgesetz bei der Gradverteilung:

```python
DFFG = generators.DynamicForestFireGenerator(0.33)
GDFF = Graph()
GUDFF = dynamic.GraphUpdater(GDFF)
GUDFF.update(DFFG.generate(999))
properties.overview(GDFF)
```

Im Gegensatz zu den anderen Geratoren ist hier die Kantenzahl zufällig:

```python
F1 = generators.DynamicForestFireGenerator(0.33)
G1 = Graph()
GU1 = dynamic.GraphUpdater(G1)
GU1.update(F1.generate(49))
m1 = G1.numberOfEdges()

F2 = generators.DynamicForestFireGenerator(0.33)
G2 = Graph()
GU2 = dynamic.GraphUpdater(G2)
GU2.update(F2.generate(49))
m2 = G2.numberOfEdges()

(m1, m2)
```
